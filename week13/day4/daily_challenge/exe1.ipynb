```python
from sklearn.cluster import KMeans

# Create a KMeans model with 4 clusters
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit the model to the data
kmeans.fit(X)

# Get cluster assignments for each data point
cluster_labels = kmeans.labels_

# Get cluster centers (centroids)
cluster_centers = kmeans.cluster_centers_

# Print the cluster centers
print("Cluster Centers:")
print(cluster_centers)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
n_samples = 300
n_clusters = 4
std_dev = 0.5

X, y = make_blobs(n_samples=n_samples, n_features=2, centers=n_clusters, cluster_std=std_dev, random_state=42)

# Create a KMeans model with 4 clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)

# Get cluster assignments for each data point
cluster_labels = kmeans.labels_

# Plot the data with predicted clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=300, c='red', label='Cluster Centers')
plt.title('Predicted Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, [2, 3]]  # Use only 'petal_length' and 'petal_width'
true_labels = iris.target

# Function to plot clusters
def plot_clusters(X, labels, centers, title):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k')
    if centers is not None:
        plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=300, c='red', label='Cluster Centers')
    plt.title(title)
    plt.xlabel('Petal Length')
    plt.ylabel('Petal Width')
    plt.colorbar()
    plt.legend()
    plt.show()

# Different values of K
k_values = [2, 3, 4, 5]

for k in k_values:
    # Create a KMeans model
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    
    # Get cluster assignments for each data point
    cluster_labels = kmeans.labels_
    
    # Get cluster centers (centroids)
    cluster_centers = kmeans.cluster_centers_
    
    # Plot the clusters
    plot_clusters(X, cluster_labels, cluster_centers, title=f'K-Means Clustering (K={k})')

# Original class distribution
plot_clusters(X, true_labels, None, title='Original Classes')
```
