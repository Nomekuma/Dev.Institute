```python
import random

data = [
    {
        "question": "What is Baby Yoda's real name?",
        "answer": "Grogu"
    },
    {
        "question": "Where did Obi-Wan take Luke after his birth?",
        "answer": "Tatooine"
    },
    {
        "question": "What year did the first Star Wars movie come out?",
        "answer": "1977"
    },
    {
        "question": "Who built C-3PO?",
        "answer": "Anakin Skywalker"
    },
    {
        "question": "Anakin Skywalker grew up to be who?",
        "answer": "Darth Vader"
    },
    {
        "question": "What species is Chewbacca?",
        "answer": "Wookiee"
    }
]

def ask_questions(questions):
    random.shuffle(questions)
    num_correct = 0
    num_questions = len(questions)
    wrong_answers = []

    for i, data in enumerate(questions, start=1):
        print(f"Question {i}/{num_questions}: {data['question']}")
        user_answer = input("Your Answer: ")

        if user_answer.lower() == data["answer"].lower():
            num_correct += 1
            print("Correct!\n")
        else:
            wrong_answers.append((data['question'], data['answer'], user_answer))
            print(f"Sorry, the correct answer is: {data['answer']}\n")

    return num_correct, num_questions - num_correct, wrong_answers

def get_feedback(num_correct, total_questions):
    percentage_correct = (num_correct / total_questions) * 100

    if percentage_correct == 100:
        return "Congratulations! You got all the answers correct. You are a Star Wars expert!"
    elif percentage_correct >= 70:
        return f"Great job! You got {num_correct}/{total_questions} answers correct. May the Force be with you!"
    else:
        return f"Nice try! You got {num_correct}/{total_questions} answers correct. Keep practicing to become a true Star Wars fan."

def main():
    print("Welcome to the Star Wars Quiz!")
    num_correct, num_incorrect, wrong_answers = ask_questions(data)
    feedback = get_feedback(num_correct, len(data))
    print(feedback)

    if num_incorrect > 0:
        print("Here are the questions you answered incorrectly:")
        for question, correct_answer, user_answer in wrong_answers:
            print(f"Question: {question}")
            print(f"Your Answer: {user_answer}")
            print(f"Correct Answer: {correct_answer}\n")

if __name__ == "__main__":
    main()
```
